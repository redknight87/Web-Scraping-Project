# Data Science Jobs & Skill Frequencies in LA

```python
#Packages
from bs4 import BeautifulSoup
import urllib
import re
import pandas as pd
import numpy as np
import requests
from urllib.request import urlopen
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
```

```python
title = []
company = []
location = []
jd = []
url = "https://www.indeed.com/m/jobs?q=data+scientist&l=Los+Angeles%2C+CA"
#url = "https://www.indeed.com/m/jobs?q=data+scientist&l=san+francisco"
for i in range(20):
    page = urlopen(url)
    soup = BeautifulSoup(page, 'lxml')
    all_matches = soup.findAll(attrs = {'rel':['nofollow']})
    
    for each in all_matches:
        jd_url = "https://www.indeed.com/m/"+each['href']
        response = requests.get(jd_url)
        jd_page = response.text
        jd_soup = BeautifulSoup(jd_page, 'html.parser')
        jd_desc = jd_soup.findAll('div', attrs = {'id':['desc']})
        
        title.append(jd_soup.body.p.b.font.text)
        company.append(jd_soup.body.p.span.previous_sibling[1:].split('-')[0])
        location.append(jd_soup.body.p.span.text)
        jd.append(jd_desc[0].text)
        
    url_all = soup.findAll(attrs = {'rel' : ['next']})
    #print(url_all)
    url = "https://www.indeed.com/m/"+str(url_all[0]['href'])

job = {'Title': title,
      'Company': company,
      'Location': location,
      'Job Description': jd}

df=pd.DataFrame.from_dict(job)
df.head(10)
```

```python
#Reads through job description and returns 'True' or 'False' if specific keyword is present
df["C"] = df["Job Description"].str.contains(pat = r'\b[Cc]\s', regex=True)
df["C++"] = df["Job Description"].str.contains(pat = r'\b[Cc]\+(?=\W)', regex=True)
df["Java"] = df["Job Description"].str.contains("Java")
df["Javascript"] = df["Job Description"].str.contains("Javascript")
df["Python"] = df["Job Description"].str.contains("Python")
df["R"] = df["Job Description"].str.contains(pat = r'\b[R]\b', regex=True)
df["SQL"] = df["Job Description"].str.contains("SQL")
df["Hadoop"] = df["Job Description"].str.contains("Hadoop")
df["Hive"] = df["Job Description"].str.contains("Hive")
df["Pig"] = df["Job Description"].str.contains("Pig")
df["Spark"] = df["Job Description"].str.contains("Spark")
df["AWS"] = df["Job Description"].str.contains("AWS")
df["Tableau"] = df["Job Description"].str.contains("Tableau")

#If keyword is present (True), append into 'Required Skills'; if False, leave blank.   
df['Required Skills'] = np.where(df['C'] == True, "C, ", " ")
df['Required Skills'] = np.where(df['C++'] == True, df['Required Skills'] + "C++, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Java'] == True, df['Required Skills'] + "Java, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Javascript'] == True, df['Required Skills'] + "Javascript, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Python'] == True, df['Required Skills'] + "Python, ", df['Required Skills'])
df['Required Skills'] = np.where(df['R'] == True, df['Required Skills'] + "R, ", df['Required Skills'])
df['Required Skills'] = np.where(df['SQL'] == True, df['Required Skills'] + "SQL, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Hadoop'] == True, df['Required Skills'] + "Hadoop, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Hive'] == True, df['Required Skills'] + "Hive, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Pig'] == True, df['Required Skills'] + "Pig, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Spark'] == True, df['Required Skills'] + "Spark, ", df['Required Skills'])
df['Required Skills'] = np.where(df['AWS'] == True, df['Required Skills'] + "AWS, ", df['Required Skills'])
df['Required Skills'] = np.where(df['Tableau'] == True, df['Required Skills'] + "Tableau, ", df['Required Skills'])

#Counts the number of required skills in the job description
df['Number of Required Skills'] = df['Required Skills'].str.split().str.len()
df = df[['Title', 'Company', 'Location', 'Job Description', 'Required Skills', 'Number of Required Skills']]
df
```

# Data Scientist Job Reviews

```python
company_name = []
reviewer = []
review_title = []
review_date = []
review_rating = []
review_description = []
review_pros = []
review_cons = []
review = []
pros = []
cons = []

url = "https://www.indeed.com/cmp?q=Data+Scientist%2FData+Analyst&l=Los+Angeles%2C+CA&from=discovery-cmp-search"
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page, "html.parser")
all_matches = soup.findAll('a', attrs = {'itemprop':['name'], 'class':['cmp-CompanyWidget-name']})

#Diving into each of the job links
for a in all_matches:
    j_url = "https://www.indeed.com"+a['href']
    response_1 = requests.get(j_url)
    j_page = response_1.text
    j_soup = BeautifulSoup(j_page, 'html.parser')
    all_matches_2 = j_soup.find_all('a', attrs={'data-tn-element':['reviews-tab'],'class':["cmp-MenuItem-link cmp-u-noUnderline"]})
    
    for each in all_matches_2:
        r_url = "https://www.indeed.com"+each['href']
        for i in range(3):
            response_2 = requests.get(r_url)
            r_page = response_2.text
            r_soup = BeautifulSoup(r_page, 'html.parser')

            '''Because of the adaptive page (i.e. class name changing randomly like 'cmp-CompanyName' and 'cmp-company-name') 
            we write if statements so we do not append any 'blanks' into our list.'''
            
            #Extracting the title of the Company
            c_title_2 = r_soup.findAll('div', attrs = {'class':['cmp-CompanyName'], 'itemprop':['name']})
            if c_title_2 == []:
                c_title_2 = r_soup.findAll('div', attrs = {'class':['cmp-company-name'], 'itemprop':['name']})
            company_name.append([x.text for x in c_title_2])

            #Extracting the Name of the Reviewer for the review
            r_container_2 = r_soup.findAll('span', attrs = {'class':['cmp-reviewer']})
            if r_container_2 == []:
                r_container_2 = r_soup.findAll('span', attrs={'class':['cmp-ReviewAuthor'],'itemprop':['author']})
            reviewer.append([y.text for y in r_container_2])

            #Extracting the Name of the Reviewer for the review
            r_date = r_soup.findAll('span',attrs={'class':['cmp-review-date-created']})
            if r_date!=[]:
                r_date=[x.text for x in r_date]
            if r_date==[]:
                r_date= r_soup.findAll('span',attrs={'class':['cmp-ReviewAuthor'],'itemprop':['author'],'itemtype':['http://schema.org/Person']})
                r_date=[y.text for y in r_date]
                for x in range(len(r_date)):
                    element=r_date[x].split('-')[-1]
                    r_date[x]=element
            review_date.append(r_date)

            #Extracting the Title for the review
            r_title = r_soup.findAll('div', attrs = {'class':['cmp-review-title']})
            if r_title == []:
                r_title = r_soup.findAll('div', attrs = {'class':['cmp-Review-title']})
            review_title.append([y.text for y in r_title])

            #Extracting the Rating for the review
            r_rating = r_soup.findAll('div', attrs = {'class':['cmp-ratingNumber']})
            if r_rating == []:
                r_rating = r_soup.findAll('div', attrs = {'class':['cmp-ReviewRating-text']})
            review_rating.append([y.text for y in r_rating])

            #Extracting the Description for the review
            r_desc = r_soup.findAll('div', attrs = {'class':['cmp-review-content-container']})
            if r_desc != []:            
                rpc = [x.text for x in r_desc]
                review_description = [None] * len(rpc)
                review_pros = [None] * len(rpc)
                review_cons = [None] * len(rpc)

                for y in range(len(rpc)):
                    review_x = rpc[y].split('Pros')[0]                
                    review_description[y]=review_x

                    pros_x = rpc[y].split('Pros')[-1].split('Cons')[0]
                    if pros_x == review_x:
                        review_pros[y] = 'NA'
                    else:
                        review_pros[y] = pros_x

                    cons_x = rpc[y].split('Pros')[-1].split('Cons')[-1]
                    if cons_x == review_x:
                        review_cons[y] = 'NA'
                    else:
                        review_cons[y] = cons_x

            else:
                r_desc = r_soup.findAll('div', attrs = {'class':['cmp-Review-text'], 'data-tn-component':['reviewDescription']})
                rpc = [x.text for x in r_desc]
                review_description = [None] * len(rpc)
                review_pros = [None] * len(rpc)
                review_cons = [None] * len(rpc)

                for y in range(len(rpc)):
                    pros_x = rpc[y].split('Pros')[-1].split('Cons')[0]
                    cons_x = rpc[y].split('Pros')[-1].split('Cons')[-1]
                    review_x = rpc[y].split('Pros')[0]

                    review_description[y]=review_x

                    if pros_x == review_x:
                        review_pros[y] = "NA"
                    else:
                        review_pros[y] = pros_x

                    if cons_x == review_x:
                        review_cons[y] = "NA"
                    else:
                        review_cons[y] = cons_x

            review.append(review_description)
            pros.append(review_pros)
            cons.append(review_cons)

            j_next = r_soup.findAll(attrs = {'rel':['next']})
            if j_next == []:
                #print('This Company does not contain more than 1 page of reviews')
                break
            else:
                if str(j_next[0]['href'])[0:4] == 'http':
                    r_url = str(j_next[0]['href'])
                else:
                    r_url = "https://www.indeed.com/"+str(j_next[0]['href'])
                #print(r_url)
```

```python
data = pd.DataFrame.from_dict({'company': company_name,
        'reviewer': reviewer,
        'review_title': review_title,
        'ratings': review_rating,
        'date': review_date,
        'descr': review,
        'pros': pros,
        'cons': cons
        })

C = []
R = []
RT = []
RR = []
RDTE = []
RDESC = []
RP = []
RC = []

for _, row in data.iterrows():
    com = row.company[0]
    for a in row.reviewer:
        C.append(com)
        R.append(a)
    for b in row.review_title:
        RT.append(b)
    for c in row.ratings:
        RR.append(c)
    for d in row.date:
        RDTE.append(d)
    for e in row.descr:
        RDESC.append(e)
    for f in row.pros:
        RP.append(f)
    for g in row.cons:
        RC.append(g)    
    
    
data_2 = pd.DataFrame({'Company': C,
                      'Reviewer': R,
                      'Review Title': RT,
                      'Review Ratings': RR,
                      'Review Date': RDTE,
                      'Review Content': RDESC,
                      'Pros': RP,
                      'Cons': RC})
data_2
```
